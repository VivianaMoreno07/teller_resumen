Importamos Librerias


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D



```

Definir el sistema de ecuaciones:

```python
A = np.array([
    [3, 9, -10],
    [1, -6, 4],
    [10, -2, 8]
])
B = np.array([24, -4, 20])
```

Resolver el sistema de ecuaciones:

```python
solucion = np.linalg.solve(A, B)
```

```python
print(f'Las soluciones del sistema son: x = {solucion[0]:.2f}, y = {solucion[1]:.2f}, z = {solucion[2]:.2f}')

# Graficar
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Definir los valores para graficar los planos
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
z = np.linspace(-10, 10, 100)

X, Y = np.meshgrid(x, y)

# Graficar los planos
ax.plot_surface(X, Y, (24 - 3*X - 9*Y) / -10, alpha=0.5, rstride=200, cstride=100)
ax.plot_surface(X, (4 - X + 6*Y) / 4, Y, alpha=0.5, rstride=200, cstride=100)
ax.plot_surface((20 - 10*X + 2*Y) / 8, Y, X, alpha=0.5, rstride=100, cstride=100)

# Graficar el punto de intersección (solución del sistema)
ax.scatter(solucion[0], solucion[1], solucion[2], color='red', s=100)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
